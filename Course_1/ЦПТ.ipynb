```python
import scipy.stats as sts
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

### Генерируем выборку объема 1000 из распределения Парето
https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.pareto.html#scipy.stats.pareto
https://ru.wikipedia.org/wiki/%D0%A0%D0%B0%D1%81%D0%BF%D1%80%D0%B5%D0%B4%D0%B5%D0%BB%D0%B5%D0%BD%D0%B8%D0%B5_%D0%9F%D0%B0%D1%80%D0%B5%D1%82%D0%BE

```python
b = 3
pareto = sts.pareto(b)
sample = pareto.rvs(1000)

EX = b / (b - 1)
DX = b / ((b - 1)** 2 * (b - 2))
```

### Строим гистограмму выборки и рисуем теоретическую плотность распределения нашей величины

```python
plt.hist(sample, normed=True, label='real')

x = np.linspace(0,10,100)
pdf = pareto.pdf(x)
plt.plot(x, pdf, label='teoretical')

plt.ylabel('fraction of samples')
plt.xlabel('$x$')
plt.legend()
plt.show()
```

## Распределение выборочного среднего при  n = 5

```python
n_samples = 1000
n = 5
means = []

for i in range(n_samples):
    mean = pareto.rvs(n).mean()
    means.append(mean)
    
means = np.array(means)
plt.hist(means, normed=True)
plt.xlabel('Выборочное среднее')
plt.ylabel('fraction of samples')
plt.title('Распределение выборочного среднего при  n = ' + str(n))

norm_rv = sts.norm(loc=EX, scale=(DX/n)**0.5)
x = np.linspace(0,4,100)
pdf = norm_rv.pdf(x)
plt.plot(x, pdf)
plt.show()
```
